```python
import numpy as np
from matplotlib import pyplot as plt
```

$$ y =tan(x)$$

```python
x=np.arange(0,4*np.pi,0.1)
x
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.tan(x)
plt.plot(x,y);
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(1)
```

```python
x=np.arange(0,4*np.pi,0.1)
y =np.cos(x)
plt.plot(x,y);
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(1)
```

$$ y =2sin(4x)-cos(3x)$$

```python
x=np.linspace(-10,10,1000)
y=np.sin(x)
z=np.cos
plt.plot(x,((2*y*4*x)-z(3*x)));
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(1)
```

$$ y =\frac{e^x}{1+e^x}$$

```python
x=np.linspace(-10,10,100)
y=np.e**x
plt.plot(x,y/(1+y));
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(1)
```
